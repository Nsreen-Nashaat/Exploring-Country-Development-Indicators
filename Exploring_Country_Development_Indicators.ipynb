```python
#Improts
 
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import sqlite3
import matplotlib.pyplot as plt
```

```python
# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory
 
database ='Country_Development_Indicators_Dataset.sqlite'
 
```

```python
# Task 1: Connect to the Database
conn = sqlite3.connect(database)
```

```python
query = """select *
           from sqlite_master
           where type='table';"""

tables = pd.read_sql(query, conn)
tables
```

```python
# Task 2: Explore the Indicators Table

Indicators = pd.read_sql("""SELECT * 
                         FROM Indicators;""", conn)
Indicators
```

```python
# Task 3: Check the Number of Rows
rows_num = pd.read_sql("""SELECT COUNT(*) AS rows_num
                         FROM Indicators;""", conn)
rows_num
```

```python
# Task4. Check for Missing Values in One Column
missing_value = pd.read_sql(""" SELECT COUNT(*) AS missing_count
                                FROM Indicators
                                WHERE Value IS NULL;""", conn)
missing_value
```

```python
#Task 5. Filter Non-Null Values
not_null_val = pd.read_sql("""SELECT COUNT(*) AS not_null_value
                            FROM Indicators
                            WHERE Value IS NOT NULL;""", conn)
not_null_val
```

```python
# Task 6. Check for Missing Values in All Columns
any_missing = pd.read_sql("""
    SELECT COUNT(*) AS rows_with_any_null
    FROM Indicators
    WHERE 
        CountryCode IS NULL OR
        IndicatorCode IS NULL OR
        Year IS NULL OR
        Value IS NULL;
""", conn)
any_missing
```

```python
# Task 7. How Many Different Indicators Are There?
indicators_num = pd.read_sql("""
    SELECT COUNT(DISTINCT IndicatorName) AS num_indicators
    FROM Indicators;
""", conn)
indicators_num
```

```python
# Task 8. Select Distinct Indicators
indicators_name = pd.read_sql("""
    SELECT DISTINCT IndicatorName AS indicators_name
    FROM Indicators;
""", conn)
indicators_name
```

```python
# Task 9. Search for an Indicator Related to GDP
gdp_indicators = pd.read_sql("""
    SELECT DISTINCT  IndicatorName
    FROM Indicators
    WHERE IndicatorName LIKE '%GDP%';
""", conn)
gdp_indicators
```

```python
# 10. Search for GDP Per Capita of Brazil in the Last 10 Years
gdp_brazil = pd.read_sql(""" SELECT *
                             From Indicators
                             WHERE CountryCode = 'BRA'
                                AND IndicatorName LIKE '%GDP%'
                                AND IndicatorName Like '%current US$%'
                                AND Year > 2004
                             ORDER BY year DESC ;  """,conn)
gdp_brazil
```

```python
# 11. Compare Brazil’s GDP Per Capita with China’s
gdp_braz_Vs_chin = pd.read_sql(""" SELECT *
                             From Indicators
                             WHERE CountryCode IN ('BRA','CHN')
                                AND IndicatorName LIKE '%GDP%'
                                AND IndicatorName Like '%current US$%'
                                AND Year > 2004
                             ORDER BY year DESC ;  """,conn)
gdp_braz_Vs_chin
```

```python
# 12. Check Brazil’s GDP Per Capita in the 1990s
gdp_brazil_1990 = pd.read_sql(""" SELECT *
                             From Indicators
                             WHERE CountryCode = 'BRA'
                                AND IndicatorName LIKE '%GDP%'
                                AND IndicatorName Like '%current US$%'
                                AND Year = 1990; """,conn)
gdp_brazil_1990
```

```python
# 13. Check Indicators for Brazil, China, and India in 2014
gdp_2014 = pd.read_sql(""" SELECT *
                             From Indicators
                             WHERE CountryCode IN ('BRA', 'CHN', 'IND')
                                AND Year = 2014; """,conn)
gdp_2014
```

*** Part 2: Aggregating and Summarizing

```python
# 1. Search for the Highest GDP Per Capita in 2014
highest_GDP  = pd.read_sql("""  SELECT CountryName, MAx(Value) AS MAX_GDP
                                From Indicators
                                WHERE IndicatorName LIKE '%GDP%'
                                AND IndicatorName Like '%current US$%'
                                AND Year = 2014; """,conn)
highest_GDP
```

```python
# 2. Reaffirm the Highest GDP Per Capita
highest_GDP_raff = pd.read_sql(""" SELECT CountryName, Value
                             From Indicators
                             WHERE  IndicatorName LIKE '%GDP%'
                                AND IndicatorName Like '%current US$%'
                                AND Year = 2014
                               group by countryName
                                ORDER BY Value DESC; """,conn)
highest_GDP_raff
```

```python
# 3. Compare the Averages of GDP Per Capita
avg_gdp = pd.read_sql(""" SELECT CountryName, AVG(Value) AS AVG_GDP_per_country
                             From Indicators
                             WHERE  IndicatorName LIKE '%GDP%'
                                AND IndicatorName Like '%current US$%'
                                AND Year = 2014
                             group by countryName 
                             ORDER BY AVG(Value) DESC; """,conn)
avg_gdp
```

```python
# 4. How Many Measures During This Time?
measures = pd.read_sql(""" SELECT Year, COUNT(*) AS measures
                             From Indicators
                             WHERE  Year > 2004
                             GROUP BY Year
                             ORDER BY Year DESC; """,conn)
measures
```

```python
# 5. Note on Angola's Data
Angola_measures = pd.read_sql(""" SELECT CountryName, Year, COUNT(*) AS measures
                             From Indicators
                             WHERE  Year > 2004
                             GROUP BY CountryName, Year
                             ORDER BY Year DESC, measures ASC; """,conn)
Angola_measures
```

```python
# 1. Join to Get GDP-Related Information in 2014 for Brazil
gdp_brazil_2014 = pd.read_sql(""" SELECT * 
                                  FROM  Country AS c 
                                  INNER JOIN Indicators AS i 
                                  ON c.CountryCode = i.CountryCode
                                  WHERE CountryName = 'Brazil'
                                     AND Year =2014;""",conn)
gdp_brazil_2014
```
